```python
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score
```

```python
train_df = pd.read_csv('/content/sample_data/train.csv')
test_df = pd.read_csv('/content/sample_data/test.csv')
```

```python
# dropped column f27 due to irrelevant string data
train_df=train_df.drop("f_27",axis=1)
test_df=test_df.drop("f_27",axis=1)
# dropped irrelevant data
train_df=train_df.drop('f_29',axis=1)
train_df=train_df.drop('f_30',axis=1)
test_df=test_df.drop('f_29',axis=1)
test_df=test_df.drop('f_30',axis=1)
print("Train data is")
print(train_df)
print("Test data is")
print(test_df)
```

```python
#dropped 'id' from the train set but not from test data
train_df=train_df.drop("id",axis=1)
# Apply Machine Learning Models
Y_train = train_df["target"]
X_train = train_df.drop("target", axis=1)
X_test  = test_df.drop("id", axis=1).copy()
print("Y Train")
print(Y_train)
print("X Train")
print(X_train)
print("X test")
print(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
svm = LinearSVC()
svm.fit(X_train, Y_train)
y_predict=svm.predict(X_test)
print(y_predict)
```

```python
print(y_predict.shape)
print(y_predict)

AsadullahCsv = pd.DataFrame({
    "id": test_df["id"],
    "target": y_predict
})
AsadullahCsv.to_csv('Asadullah_svm_submission.csv', index=False)
```

```python
scores = cross_val_score(svm, X_train, Y_train, cv=4, scoring = "accuracy is")
print("Scores is:\n", pd.Series(scores))
print("Mean is:", scores.mean())
print("Standard Deviation is:", scores.std())
```
